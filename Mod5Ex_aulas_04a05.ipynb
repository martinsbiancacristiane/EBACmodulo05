#### Marcação de bom e mau
O objetivo da modelagem é classificar o risco de inadimplência, ou como se diz no meio, o risco de *default*. Podemos fazer longas discussões sobre o conceito de *default* com base em estudos e exigências regulatórias, para efeitos deste estudo, um cliente em *default* é aquele que está em 60 dias de atraso ou mais. Então classificaremos os clientes como 'bons' e 'maus' assim:
- **Maus** pagadores: são aqueles que entraram em 'default' (atraso 60 dias ou mais) nos 24 meses seguintes à aquisição do cartão de crédito. 
- **Bons** pagadores: são considerados todos os demais.
- **Excluídos**: Clientes que não adquiriram um cartão de crédito (seja por recusa, seja por desistência) não possuem informações de pagamento, portanto não se pode identificar se são bons ou maus. Há uma longa discussão e literatura sobre *inferência de rejeitados* que está fora do escopo deste exercício.

#### Construindo a variável resposta
A base de pagamentos está em um formato de 'base larga'. Essa base possui informações de pagamentos do cliente mês a mês a partir do mês de aquisição do crédito (mês 0) até o vigésimo quarto mês após a aquisição do crédito (mês 24). Utilizaremos essa base para determinar se um proponente é considerado 'bom pagador' ou caso apresente atraso representativo, será considerado 'mau pagador'.

#### Base larga vs base longa
A base ser larga significa que há uma linha para cada cliente, e que as informações estarão nas colunas, em contraste com a 'base longa', em que haveria uma linha para cada combinação cliente/mês, uma coluna indicando o cliente, outra indicando o mês, e apenas uma coluna com a informação do atraso.

````python
import pandas as pd 

propostas = pd.read_csv('application_record.csv')
pg = pd.read_csv('pagamentos_largo.csv')
````

````python

# --- Passo 1: Definir os códigos que representam DEFAULT ---
# Atraso de 60 dias ou mais ('2', '3', '4', '5')
codigos_default = ['2', '3', '4', '5']

# --- Passo 2: Isolar as colunas de mês ---
# Filtra todas as colunas que começam com 'mes_'
colunas_mes = [col for col in pg.columns if 'mes_' in col]

# --- Passo 3: Aplicar o .isin() para criar o indicador booleano ---
# O .isin() verifica cada célula nas colunas de mês, gerando True/False.
pg_default_mensal = pg[colunas_mes].isin(codigos_default)

# --- Passo 4: Opcional, mas útil: Anexar o ID para referência futura ---
# Adicionamos a coluna 'ID' de volta ao novo DataFrame de indicadores.
pg_default_mensal.insert(0, 'ID', pg['ID'])

print("Tarefa 1 CONCLUÍDA: DataFrame de Indicadores Mensais (pg_default_mensal)")
print(pg_default_mensal.head())
````

````python
# 1. Remove a coluna 'ID' do DataFrame de indicadores para somar APENAS os meses
# Usa .drop('ID', axis=1)
pg_mensal_soma = pg_default_mensal.drop('ID', axis=1)

# 2. Soma o número de meses em default para cada cliente (axis=1 = soma horizontal)
# O resultado é uma nova coluna no DataFrame 'pg'
pg['meses_em_default'] = pg_mensal_soma.sum(axis=1)

# 3. Cria a variável resposta final 'mau_pagador'
# True se o número de meses em default for MAIOR OU IGUAL a 1
pg['mau_pagador'] = (pg['meses_em_default'] >= 1)

# Verificação (para cumprir o requisito de checagem)
print("Contagem de Maus Pagadores (na base de pagamentos):")
print(pg['mau_pagador'].value_counts())
````

````python
# TAREFA 3: Marcando proponentes expostos ao risco de crédito


# 1. Cria uma Série (lista) dos IDs presentes na base de pagamentos (os tomadores)
ids_tomadores = pg['ID']

# 2. Filtra o DataFrame 'propostas' usando a lista de IDs de tomadores
# O resultado é um novo DataFrame contendo apenas os proponentes que viraram tomadores.
propostas_tomadores = propostas[propostas['ID'].isin(ids_tomadores)]

# Verificação (para checar o sucesso do filtro)
print(f"Total de proponentes originais: {len(propostas)}")
print(f"Total de clientes tomadores (filtrados): {len(propostas_tomadores)}")
````

````python
# TAREFA 4: Consolidando as informações (Junção das bases)
# ===================================================================================

# 1. Prepara a base da variável resposta (apenas o ID e a variável alvo)
df_resposta = pg[['ID', 'mau_pagador']]

# 2. Executa o Inner Join entre Propostas e a Resposta.
# O 'inner' garante que só os clientes que se tornaram tomadores sejam mantidos.
df_final = pd.merge(
    propostas,       # Tabela da Esquerda (contém todas as variáveis preditoras)
    df_resposta,     # Tabela da Direita (contém a variável alvo)
    on='ID',         # A chave de ligação
    how='inner'      # O tipo de junção (apenas correspondências em ambas as tabelas)
)

# Confirmação do novo tamanho e colunas
print("DataFrame Final Criado (df_final)")
print(f"Total de linhas consolidadas: {len(df_final)}")
print("Colunas do DataFrame final (amostra):", df_final.columns.tolist()[:5], ['...', 'mau_pagador'])
````
